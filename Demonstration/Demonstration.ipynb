```python
# Display real image and the respective style transfreed image

%matplotlib inline
fig, axes = plt.subplots(ncols=5, sharex='all',sharey='all',figsize=(13,13))
img = mpimg.imread('./datasets/avg/download.png')
axes[0].imshow(img)
axes[0].set_title('Real')
img = mpimg.imread('./results/style_cezanne_pretrained/test_latest/images/download_fake_B.png')
axes[1].imshow(img)
axes[1].set_title('Cezanne')
img = mpimg.imread('./results/style_monet_pretrained/test_latest/images/download_fake_B.png')
axes[2].imshow(img)
axes[2].set_title('Monet')
img = mpimg.imread('./results/style_ukiyoe_pretrained/test_latest/images/download_fake_B.png')
axes[3].imshow(img)
axes[3].set_title('Ukiyoe')
img = mpimg.imread('./results/style_vangogh_pretrained/test_latest/images/download_fake_B.png')
axes[4].imshow(img)
axes[4].set_title('Vangogh')
plt.show()
```
